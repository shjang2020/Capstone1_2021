```python
import matplotlib.font_manager as fm

from matplotlib import rc

font_name = fm.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()

rc('font', family=font_name)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 8))

nodes = {
    "후기": dict(color="lightsalmon"),
    "이사": dict(color="peachpuff"),
    "스튜디오": dict(color= 'peachpuff'),
    '유아용품' : dict(color = "lightsalmon"),
    '유모차' : dict(color = 'peachpuff'),
    '책상' : dict(color = 'peachpuff'),
    '침대' : dict(color = 'peachpuff'),
    '시트' : dict(color = 'peachpuff'),
    '인테리어' : dict(color = 'peachpuff'),
    '커튼' : dict(color = 'peachpuff'),
    '가구' : dict(color = "lightsalmon"),
    '소파' : dict(color = 'peachpuff'),
    '청소기' : dict(color = 'peachpuff'),
    '다이어트' : dict(color = "lightsalmon"),
    '카시트' : dict(color = "lightsalmon"),
    '아이' : dict(color = 'peachpuff'),
    '사이즈' : dict(color = 'peachpuff'),
    '층간소음' : dict(color = "lightsalmon"),
    '매트' : dict(color = 'peachpuff'),
    '시공' : dict(color = 'peachpuff'),
    '블럭' : dict(color = 'peachpuff'),
}
edges = [
    ("후기", "이사", 'Weak', ),
    ("후기", "스튜디오", "Weak"),
    ("유아용품", "유모차", "Weak"),
    ("유아용품", "책상", "Weak"),
    ("유아용품", "침대", "Weak"),
    ("유아용품", "시트", "Weak"),
    ("유아용품", "인테리어", "Weak"),
    ("유아용품", "커튼", "Weak"),
    ("가구", "소파", "Weak"),
    ("가구", "청소기", "Weak"),
    ("가구", "침대", "Weak"),
    ("가구", "시트", "Weak"),
    ("가구", "책상", "Weak"),
    ("카시트", "아이", "Weak"),
    ("카시트", "사이즈", "Weak"),
    ("층간소음", "매트", "Weak"),
    ("층간소음", "시공", "Weak"),
    ("층간소음", "블럭", "Weak"),
]
original_graph = nx.Graph()
original_graph.add_nodes_from(n for n in nodes.items())
original_graph.add_edges_from((u, v, {"type": label}) for u, v, label in edges)


plt.suptitle("키워드 내돈내산 제목 주제분석")

base_options = dict(with_labels=True, edgecolors = "black", node_size = 2000, font_family = font_name)

plt.title(
    "(%s 토픽, %s )"
    % (original_graph.number_of_nodes(), original_graph.number_of_edges())
)
pos = nx.spring_layout(original_graph)
node_colors = [d["color"] for _, d in original_graph.nodes(data=True)]

edge_type_visual_weight_lookup = {"Weak": 3.0, "Strong": 800.0}
edge_weights = [
    edge_type_visual_weight_lookup[d["type"]]
    for _, _, d in original_graph.edges(data=True)
]

nx.draw_networkx(
    original_graph, pos=pos, node_color=node_colors, width=edge_weights, **base_options
)

plt.tight_layout()
plt.show()
```
